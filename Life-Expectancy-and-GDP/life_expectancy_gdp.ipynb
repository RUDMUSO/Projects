## 1. Data preparation and preliminary data review.

### A) Import all necessary libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### B) Loading and viewing the data.

```python
data_tab = pd.read_csv('all_data.csv')
data_tab.head()
```

```python
data_tab.rename(columns={'Life expectancy at birth (years)':'LEAB'}, inplace=True)
data_tab.info()
```

Great! It looks like the data is in good shape - there are no cells with missing values, and all data types are selected correctly.

We can start the data analysis!

## 2. Data analysis.

### A) To begin, let's check if the life expectancy in different countries differs in the year 2015. And also, let's check if their GDP differs in the same year.

```python
plt.subplot(1,2,1)
sns.barplot(data=data_tab[data_tab.Year == 2015], x='Country', y='LEAB')
plt.xticks(rotation=90, fontsize=10)
plt.subplot(1,2,2)
sns.barplot(data=data_tab[data_tab.Year == 2015], x='Country', y='GDP')
plt.xticks(rotation=90, fontsize=10)
plt.subplots_adjust(wspace=0.3)
plt.show()
plt.clf()
#Based on this data, it is difficult to say anything about the nature of the relationship.
```

### B) Let's see how life expectancy has changed in each country depending on GDP.

```python


sns.lineplot(data=data_tab, x='GDP', y='LEAB', hue='Country')
plt.show()
plt.clf()
```

### C) Let's plot a scatter plot and see how the values are distributed on the axes of life expectancy and GDP.

```python
sns.scatterplot(data=data_tab, x='GDP', y='Life expectancy at birth (years)')
plt.show()
plt.clf()
#
```
