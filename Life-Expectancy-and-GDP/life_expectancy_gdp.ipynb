## 0. Definition of goals and objectives of the project.
### A) Project Goals - research on the expected life expectancy and GDP of a country and identifying the relationship between them.
### B) Data - for this analysis, data on the GDP of different countries, life expectancy, and changes in this information over time may be required. All this information is available in all_data.csv.
### C) Analysis - it will be necessary to study correlations both in the overall set of countries and for a particular country over time.

## 1. Data preparation and preliminary data review.

### A) Import all necessary libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### B) Loading and viewing the data.

```python
data_tab = pd.read_csv('all_data.csv')
data_tab.head()
```

```python
data_tab.rename(columns={'Life expectancy at birth (years)':'LEAB'}, inplace=True)
data_tab.info()
```

Great! It looks like the data is in good shape - there are no cells with missing values, and all data types are selected correctly.

We can start the data analysis!

## 2. Data analysis.

### A) To begin, let's check if the life expectancy in different countries differs in the year 2015. And also, let's check if their GDP differs in the same year.

```python
plt.subplot(1,2,1)
sns.barplot(data=data_tab[data_tab.Year == 2015], x='Country', y='LEAB')
plt.xticks(rotation=90, fontsize=10)
plt.subplot(1,2,2)
sns.barplot(data=data_tab[data_tab.Year == 2015], x='Country', y='GDP')
plt.xticks(rotation=90, fontsize=10)
plt.subplots_adjust(wspace=0.3)
plt.show()
plt.clf()
#Based on this data, it is difficult to say anything about the nature of the relationship.
```

### B) Let's observe how life expectancy has changed in each country over the years.

```python
sns.pointplot(data=data_tab, x='Year', y='LEAB', hue='Country')
plt.xticks(rotation=30)
plt.show()
plt.clf()
```

### C) Let's observe how GDP has changed in each country over the years.

```python
sns.pointplot(data=data_tab, x='Year', y='GDP', hue='Country')
plt.xticks(rotation=30)
plt.show()
plt.clf()
```

### D) We can see certain issues with the scale. Let's look at the graphs separately for each country.


```python
for column in data_tab.Country.unique(): 
    plt.figure(figsize=(12,6)) 
    plt.subplot(1,2,1)
    sns.pointplot(data=data_tab[data_tab.Country==column], x='Year', y='GDP', hue='Country')
    plt.title(f'GDP changing in {column}')
    plt.xticks(rotation=30)
    plt.subplot(1,2,2)
    sns.pointplot(data=data_tab[data_tab.Country==column], x='Year', y='LEAB', hue='Country')
    plt.title(f'LEAB changing in {column}')
    plt.xticks(rotation=30)
    plt.show()
    plt.clf()
```

On the graphs, GDP and life expectancy behave in a similar manner.

## 3. Let's try to identify correlations.

### A) Let's plot a scatter plot and see how the values are distributed on the axes of life expectancy and GDP.

```python
sns.scatterplot(data=data_tab, x='GDP', y='LEAB', hue='Country')
plt.show()
plt.clf()
```

As with the previous graphs, it is necessary to plot separate graphs for each country.

### B) Let's plot these scatter plots separately for each country. 

```python
for column in data_tab.Country.unique(): 
    sns.scatterplot(data=data_tab[data_tab.Country==column], x='GDP', y='LEAB', hue='Country')
    plt.title(f'LEAB vs GDP for {column}')
    plt.xticks(rotation=30)
    plt.show()
    plt.clf()
```

Based on the graphs, it can be observed that countries have different shapes of the dependency curve. However, in general, a positive correlation can be seen for all countries.
